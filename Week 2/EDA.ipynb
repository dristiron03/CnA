```python
import datetime
from datetime import date
day_names = []
for i in range(len(df)):
    year = df['Year'][i]
    month = df['Months'][i]
    day = df['Date'][i].split('/')[1]
    day_name = datetime.date(int(year), int(month), int(day))
    day_names.append(day_name.strftime("%A"))
day_names
```

```python
df['day'] = day_names
df
```
