### Linear Regresssion(least square)

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)
lr = LinearRegression().fit(X_train,y_train)
```

```python
print(f'lr.coef_:{lr.coef_}')
print(f'lr.intercept_:{lr.intercept_}')
```

```python
from sklearn.metrics import accuracy_score
#X_pred_train = lr.predict(X_train)
print(f'Training Set Score: {lr.score(X_train,y_train)}')
```

```python
print(f'Test set score: {lr.score(X_test, y_test)}')
```

### Ridge Regularization

```python
from sklearn.linear_model import Ridge
ridge = Ridge().fit(x_train, y_train)
print(f'Train set score:{ridge.score(x_train, y_train)}')
print(f'Test set score:{ridge.score(x_test,y_test)}')
```

```python
for i in range(1,11):
    ridge = Ridge(alpha=i).fit(x_train, y_train)
    print(f'Train set score:{ridge.score(x_train, y_train)}')
    print(f'Test set score:{ridge.score(x_test,y_test)}')
```

```python
ridge = Ridge(alpha = 0.1).fit(x_train, y_train)
print(f'Train set score:{ridge.score(x_train, y_train)}')
print(f'Test set score:{ridge.score(x_test,y_test)}')
```

```python
ridge = Ridge(alpha = 15).fit(x_train, y_train)
print(f'Train set score:{ridge.score(x_train, y_train)}')
print(f'Test set score:{ridge.score(x_test,y_test)}')
```

```python
ridge = Ridge(alpha = 50).fit(x_train, y_train)
print(f'Train set score:{ridge.score(x_train, y_train)}')
print(f'Test set score:{ridge.score(x_test,y_test)}')
```

### Lasso Regularization

```python
from sklearn.linear_model import Lasso
import numpy as np
lasso = Lasso().fit(x_train, y_train)
print("Training set score: {}".format(lasso.score(x_train, y_train)))
print("Test set score: {}".format(lasso.score(x_test, y_test)))
print("Number of features used: {}".format(np.sum(lasso.coef_ != 0)))
```

```python
lasso001 = Lasso(alpha = 0.01, max_iter=100000).fit(x_train, y_train)
print("Training set score: {}".format(lasso.score(x_train, y_train)))
print("Test set score: {}".format(lasso.score(x_test, y_test)))
print("Number of features used: {}".format(np.sum(lasso.coef_ != 0)))
```

```python
lasso00001 = Lasso(alpha = 0.0001, max_iter=100000).fit(x_train, y_train)
print("Training set score: {}".format(lasso.score(x_train, y_train)))
print("Test set score: {}".format(lasso.score(x_test, y_test)))
print("Number of features used: {}".format(np.sum(lasso.coef_ != 0)))
```
